```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## Observations:
1. when rating is between 4.5 to 4.9 it indicates excellent
2. when rating is between 4.0 to 4.5 it indicates very good
3. when rating is between 3.5 to 3.9 it indicates good
4. when rating is between 3.0 to 3.4 it indicates average
5. when rating is between 2.5 to 9.9 it indicates average
6. when rating is between 2.0 to 2.4 it indicates poor
7. 2148 people have not rated
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings, palette=['grey','red','orange','yellow','green','green'])
```

```python
Observation:
1. Not Rated count is very high
2. Maximum  number of ratings are between 2.5 to 3.4 
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['grey','red','orange','yellow','green','green'])
```
